## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img_in = plt.imread('data/astronaut.png')
plt.figure(figsize= (5,5))
plt.imshow(img_in, 'grey')
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        if p[0] - n < 0 or p[0] + n >= img_in.shape[0] or p[1] - n < 0 or p[1] + n >= img_in.shape[1]:
            continue 
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            di = q[0] - n
            dj = q[1] - n

            ni = p[0] + di
            nj = p[1] + dj

            r = img_in[ni,nj]
            neighborhood_sum += r
        result[p] = neighborhood_sum/(size*size)
    return result
```

```python
imgfilter3= meanfilter(img_in, 3)
plt.imshow(imgfilter3, 'grey')
```

**<a id="#">Task 3.1.4.</a>** Compare your result for filter size `5` with the correct result image `data/astronaut_meanfilter5.png`:

```python
imgfilter5= meanfilter(img_in, 5)
plt.imshow(imgfilter5, 'grey')
```

```python
imgfilter5correct = plt.imread('data/astronaut_meanfilter5.png')
plt.imshow(imgfilter5correct, 'grey')
```
